```python
a = np.array([[10, 20, 30], [40, 50, 60, 70]])
print(a, type(a))

a[1][0]
```

```python
a = np.array([[10, 20, 30, 40], ["50", 60, 70, 80]])
a
```

```python
print(a.dtype)
print(a.size)     # Number of items in the array
print(a.shape)
print(a.ndim)
print(a.itemsize)  # Size (in bytes) of each element
print(a.nbytes)    # Total size (in bytes) of all items in the array
print(a.strides)
```

```python
a = np.array([[10, 20, 30], [40, 50, 60, 70, 80]], dtype=object)
print(a, type(a))
print(a[1][1])# cannot use this with object -> a[1, 1])
print(a[1])
print("-----")
print(a.shape)
print(a.dtype)
print(a.size)
print(a.ndim)
print(a.itemsize)
print(a.nbytes) # size in bytes for each item size
print(a.strides)
```
